```python

from google.colab import files
uploaded = files.upload()


import base64

filename = next(iter(uploaded))  # Get uploaded file name
with open(filename, "rb") as image_file:
    base64_image = base64.b64encode(image_file.read()).decode("utf-8")


import requests
import json

url = "" #url endpoint here

headers = {
    "Api-Key": "",  #  Replace this with your actual API key
    "Content-Type": "application/json"
}

payload = {
    "images": [base64_image],
    "similar_images": True,
    "latitude": -1.286389,       # Optional: Nairobi latitude
    "longitude": 36.817223,      # Optional: Nairobi longitude
    "datetime": "2025-05-08T10:00"  # Optional: Date/time image was taken
}

response = requests.post(url, headers=headers, data=json.dumps(payload))


from IPython.display import HTML

try:
    result = response.json()

    # Extract relevant information
    image_url = result['input']['images'][0]  # URL of uploaded image
    disease_name = result['result']['disease']['suggestions'][0]['name']
    disease_probability = result['result']['disease']['suggestions'][0]['probability']

    # Get plant name (if available)
    plant_name = "Unknown"  # Default value if plant name not found
    if 'crop' in result['result'] and 'suggestions' in result['result']['crop'] and len(result['result']['crop']['suggestions']) > 0:
        plant_name = result['result']['crop']['suggestions'][0]['name']


    # Create HTML structure
    html_content = f"""
    <h2>Crop Diagnosis</h2>
    <div style="display: flex;">
        <img src="{image_url}" width="300" height="200" alt="Uploaded Image">
        <div style="margin-left: 20px;">
            <h3>Plant Name: {plant_name}</h3>
            <h3>Likely Disease: {disease_name}</h3>
            <p>Probability: {disease_probability:.2%}</p>
        </div>
    </div>
    """

    # Display HTML
    display(HTML(html_content))

except Exception as e:
    print("Failed to parse response:", e)
    print(response.text)
```
